```python
import pandas as pd
from string import capwords
```

```python
df = pd.read_csv('/kaggle/input/openaddresses-us-northeast/ma.csv',
                 dtype={'NUMBER': str, 'POSTCODE': str})
df = df[df['CITY'] == 'BOSTON']
df
```

```python
df['address'] = df['NUMBER'] + ' ' + \
    df['STREET'].apply(capwords) + ', ' + \
    df['CITY'].apply(capwords) + ', ' + df['REGION']
dropped_df = df.drop(['LON', 'LAT', 'NUMBER', 'STREET', 'CITY', 'REGION',
                      'UNIT', 'DISTRICT', 'ID', 'HASH'], axis=1, inplace=False)
dropped_df.rename(columns={'POSTCODE': 'postcode'}, inplace=True)
dropped_df.drop_duplicates(inplace=True)
dropped_df
```

```python
with open(f'/kaggle/working/boston-addresses.jsonl', 'w') as f:
    json = dropped_df.to_json(orient='records', lines=True)
    f.write(json)
```
